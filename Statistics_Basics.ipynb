###1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.

ANS:-**Qualitative Data:-** This type of data describes qualities or characteristics. It's non-numerical and often categorical.

**Examples:-**

Colors (e.g., red, blue, green)

Gender (e.g., male, female)

Types of fruit (e.g., apple, banana, orange)

Opinions expressed in text format

**Quantitative Data:-** This type of data deals with numbers and things that can be measured objectively.

**Examples:-**

Height (e.g., 175 cm)

Weight (e.g., 68 kg)

Temperature (e.g., 25°C)

Number of students in a class (e.g., 30)

**Measurement Scales:-** The four measurement scales often used to categorize data, especially quantitative data:

**1.Nominal Scale:-** This is the most basic level. It involves categories with no inherent order or ranking.

**Examples:-**

Gender (male, female)

Blood type (A, B, AB, O)

Country of origin (USA, Canada, Mexico)

**2.Ordinal Scale:-** This scale introduces order or ranking to the categories, but the differences between them are not necessarily equal or meaningful.

**Examples:-**

Educational level (high school, bachelor's, master's, PhD)

Customer satisfaction ratings (very dissatisfied, dissatisfied, neutral, satisfied, very satisfied)

Socioeconomic status (low, middle, high)

**3.Interval Scale:-** In this scale, the differences between values are meaningful and equal, but there is no true zero point.

**Examples:-**

Temperature in Celsius or Fahrenheit (0°C or 0°F doesn't mean there's no temperature)

IQ scores (a score of 0 doesn't mean there's no intelligence)

Dates on a calendar

**4.Ratio Scale:-** This is the most advanced scale. It has all the properties of the interval scale, plus a true zero point, which indicates the absence of the quantity being measured.

**Examples:-**

Height (0 cm means no height)

Weight (0 kg means no weight)

Age (0 years old means no age)

Income (0 income means no income)



###2.What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

ANS:- **Measures of Central Tendency:-** In statistics, measures of central tendency are used to describe the "center" or typical value of a dataset. The three most common measures are:-

**1.Mean:-** The average of all the values in the dataset. It's calculated by summing all the values and dividing by the total number of values.

**Formula:-** Mean = (Sum of all values) / (Number of values)

**Example:-** The mean of the dataset [2, 4, 6, 8, 10] is (2 + 4 + 6 + 8 + 10) / 5 = 6.

**When to use:-** The mean is appropriate for datasets that are normally distributed (symmetrical) and don't have extreme outliers. It's often used for continuous data.

**2.Median:-** The middle value in a dataset that is ordered from least to greatest. If there's an even number of values, the median is the average of the two middle values.

**Example:-** The median of the dataset [2, 4, 6, 8, 10] is 6.

The median of the dataset [2, 4, 6, 8] is (4 + 6) / 2 = 5.

**When to use:-** The median is robust to outliers, meaning it's less affected by extreme values. It's often used for skewed data or data with outliers. It's also suitable for ordinal data.

**3.Mode:-** The value that appears most frequently in a dataset. A dataset can have one mode, more than one mode (multimodal), or no mode at all.

**Example:-** The mode of the dataset [2, 4, 4, 6, 8, 10] is 4.

**When to use:-** The mode is useful for categorical data or discrete data where you want to identify the most common category or value. It can also be used for nominal data.

**Situations and Examples**

Here are some examples of situations where each measure of central tendency is appropriate:

**Mean:-** Calculating the average income of a population, the average temperature of a city, or the average score on a test.

**Median:-** Analyzing house prices in a neighborhood (where outliers like mansions can skew the mean), reporting typical salaries (where a few very high earners can distort the average), or summarizing survey responses on a Likert scale.

**Mode:-** Determining the most popular product in a store, the most common blood type in a population, or the most frequent response to a survey question.



###3.  Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

ANS:- **Dispersion in Statistics:-** Dispersion in statistics refers to the spread or variability of data points in a dataset. It's a measure of how much the individual data points deviate from the central tendency (mean, median, or mode).

**High dispersion:-** Data points are widely scattered around the central tendency.

**Low dispersion:-** Data points are clustered closely around the central tendency.

**Variance and Standard Deviation:-** Variance and standard deviation are two key measures of dispersion. They quantify the spread of data by calculating the average squared distance of each data point from the mean.

**1. Variance:-**

**Formula:-** Variance = (Sum of squared differences from the mean) / (Number of data points - 1)

It's a measure of how much the data points are spread out from the mean. A higher variance indicates greater dispersion.

The variance is expressed in squared units, which can be difficult to interpret directly.

**2. Standard Deviation:-**

**Formula:-** Standard Deviation = Square root of the variance

It's the most commonly used measure of dispersion. It represents the average distance of each data point from the mean.

The standard deviation is expressed in the same units as the original data, making it easier to understand.

**Relationship between Variance and Standard Deviation:-**

Standard deviation is simply the square root of the variance.

They both measure the same thing (dispersion), but standard deviation is more commonly used because it's easier to interpret.

**How They Measure Spread**

**Variance:-** By calculating the average squared distance of each data point from the mean, the variance provides a measure of how much the data is spread out. Larger squared differences contribute more to the variance, indicating greater dispersion.

**Standard Deviation:-** Taking the square root of the variance transforms it back to the original units of the data, making it more interpretable. The standard deviation represents the typical or average distance of a data point from the mean.

**Example**

 we have two datasets:

Dataset A: [2, 4, 6, 8, 10]

Dataset B: [1, 5, 6, 7, 11]

Both datasets have the same mean (6), but Dataset B has a larger variance and standard deviation, indicating that its data points are more spread out from the mean.

In summary Dispersion measures the spread of data. Variance and standard deviation are two common measures of dispersion, with standard deviation being more widely used due to its interpretability. They quantify the spread of data by calculating the average distance of data points from the mean.

###4.  What is a box plot, and what can it tell you about the distribution of data?

ANS:- A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the five-number summary of the data:

**1.Minimum:-** The smallest value in the dataset.

**2.First Quartile (Q1):-** The value that separates the lowest 25% of the data from the rest.

**3.Median (Q2):-** The middle value of the dataset.

**4.Third Quartile (Q3):-** The value that separates the lowest 75% of the data from the rest.

**5.Maximum:-** The largest value in the dataset.

The box plot visually represents these values using a box and whiskers:

**Box:-** The box spans from Q1 to Q3, representing the interquartile range (IQR), which contains the middle 50% of the data.

**Line inside the box:-** Represents the median.

**Whiskers:-** Extend from the box to the minimum and maximum values, or to 1.5 times the IQR beyond the box (to identify potential outliers).

**Outliers:-** Data points that fall outside the whiskers are plotted as individual points and may be considered outliers.

Box plots provide valuable insights into the distribution of data, including:-

**1.Central Tendency:-** The median (line inside the box) shows the central value of the data.

**2.Dispersion:-** The IQR (length of the box) indicates the spread of the middle 50% of the data. Longer boxes suggest greater variability.

**3.Skewness:-** The position of the median within the box and the lengths of the whiskers can reveal skewness:

  **Symmetrical Distribution:-** The median is in the center of the box, and the whiskers are roughly equal in length.

  **Right-Skewed Distribution:-** The median is closer to Q1, and the right whisker is longer than the left whisker.

  **Left-Skewed Distribution:-** The median is closer to Q3, and the left whisker is longer than the right whisker.

**4.Outliers:-** Outliers are plotted as individual points beyond the whiskers, highlighting potential extreme values.

**Example**
Consider a box plot of exam scores. The box might span from 60 to 80, with the median at 70. This tells us that the middle 50% of students scored between 60 and 80, with a typical score of 70. If the right whisker is longer, it suggests that there are a few students who scored significantly higher than the rest, indicating a right-skewed distribution.

**Benefits of Box Plots**

**Easy to understand:-** Box plots provide a clear visual summary of the data distribution.

**Comparison:-** Multiple box plots can be placed side-by-side to compare distributions across different groups or categories.

**Outlier detection:-** Box plots help identify potential outliers, which can be further investigated.

###5. Discuss the role of random sampling in making inferences about populations.

ANS:- **Random Sampling and Population Inferences**

In statistics, we often want to make inferences about a population based on a sample. A population is the entire group of individuals, objects, or events that we're interested in studying, while a sample is a smaller subset of the population that we actually collect data from.

Random sampling is a fundamental technique for selecting a sample that is representative of the population. It ensures that every member of the population has an equal chance of being selected for the sample. This is crucial for making valid inferences about the population based on the sample data.

**Why is Random Sampling Important?**

**1.Avoiding Bias:-** Random sampling helps to avoid bias, which is a systematic error that can distort the results of a study. If a sample is not randomly selected, it may not accurately reflect the characteristics of the population, leading to inaccurate inferences.

**2.Representativeness:-** Random sampling increases the likelihood that the sample will be representative of the population. This means that the sample will have similar characteristics to the population, allowing us to generalize our findings from the sample to the population.

**3.Inference Validity:-** Random sampling is essential for the validity of statistical inferences. When we use random sampling, we can use probability theory to estimate the uncertainty associated with our inferences. This allows us to quantify how confident we are that our findings from the sample apply to the population.

**How Random Sampling Enables Inferences**

By ensuring a representative sample, random sampling allows us to estimate population parameters (such as the mean or standard deviation) based on sample statistics (such as the sample mean or sample standard deviation).

We can use statistical tests to determine whether differences observed in the sample are likely to reflect real differences in the population.

Confidence intervals can be calculated to provide a range of values within which we are confident the true population parameter lies.

**Example**

Suppose we want to estimate the average height of all students in a school. We could randomly select a sample of 100 students and measure their heights. Because the sample was randomly selected, we can use the sample mean height to estimate the average height of all students in the school. We can also calculate a confidence interval to quantify the uncertainty associated with our estimate.

**Conclusion**

Random sampling is essential for making valid inferences about populations based on samples. It helps to avoid bias, ensure representativeness, and enable the use of statistical methods to quantify the uncertainty associated with our inferences. By using random sampling, we can confidently generalize our findings from the sample to the larger population of interest.

###6.  Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

ANS:- **Skewness in Statistics:-** Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. In simpler terms, it describes the shape of a distribution and how it deviates from a perfectly symmetrical bell curve (normal distribution).

**Types of Skewness:-** There are three main types of skewness:-

**1.Right Skewness (Positive Skewness):-**

The tail of the distribution extends longer on the right side.

The mean is typically greater than the median, which is greater than the mode.

This indicates that there are a few unusually high values in the dataset that are pulling the mean towards the right.

**2.Left Skewness (Negative Skewness):-**

The tail of the distribution extends longer on the left side.

The mean is typically less than the median, which is less than the mode.

This indicates that there are a few unusually low values in the dataset that are pulling the mean towards the left.

**3.Zero Skewness (Symmetrical Distribution):-**

The distribution is perfectly symmetrical, with the mean, median, and mode all being equal.

This is the ideal scenario for many statistical analyses.

**How Skewness Affects Data Interpretation:-** Skewness can significantly affect the interpretation of data in several ways:

**1.Central Tendency:-** In skewed distributions, the mean may not be the best measure of central tendency, as it can be heavily influenced by extreme values. The median is often a more robust measure in these cases.

**2.Dispersion:-** Skewness can affect the interpretation of dispersion measures like standard deviation. In skewed distributions, the standard deviation may not accurately reflect the spread of the data, especially in the presence of outliers.

**3.Statistical Tests:-** Many statistical tests assume a normal distribution. If the data is skewed, these tests may not be valid, and alternative methods may need to be used.

**4.Data Transformations:-** Skewed data can often be transformed to make it more symmetrical, which can improve the validity of statistical analyses. Common transformations include logarithmic, square root, and reciprocal transformations.

**Examples of Skewness and Interpretation**

**Right Skewed:-** Income distribution is often right-skewed, with a few high earners pulling the mean upwards.

**Left Skewed:-** Exam scores might be left-skewed if a few students perform poorly, lowering the mean.

**Symmetrical:-** Heights of adult females tend to follow a symmetrical distribution.

**Conclusion**

Skewness is an important concept in statistics that describes the shape of a distribution. Understanding skewness is crucial for interpreting data correctly, as it can affect the choice of central tendency measures, dispersion measures, and statistical tests. By recognizing and addressing skewness, we can ensure that our data analyses are accurate and reliable.

###7.  What is the interquartile range (IQR), and how is it used to detect outliers?

Ans:- The interquartile range (IQR) is a measure of statistical dispersion, specifically the spread of the middle 50% of the data. It's calculated as the difference between the third quartile (Q3) and the first quartile (Q1):

**IQR = Q3 - Q1**

**Q1:** The value that separates the lowest 25% of the data from the rest.

**Q3:-** The value that separates the lowest 75% of the data from the rest.

**How is IQR Used to Detect Outliers?**

The IQR is used to establish "fences" or thresholds beyond which data points are considered potential outliers. The fences are calculated as follows:

**Lower Fence: Q1 - 1.5 * IQR**

**Upper Fence: Q3 + 1.5 * IQR**

Any data point that falls below the lower fence or above the upper fence is flagged as a potential outlier.

**Reasoning:-** The IQR method for outlier detection is based on the idea that data points that are significantly far from the central 50% of the data are likely to be unusual or extreme values. By setting the fences at 1.5 times the IQR, we create a range that encompasses the majority of the data, while identifying potential outliers that fall outside this range.

**Example:-**

Let's say we have a dataset of exam scores with Q1 = 60, Q3 = 80, and IQR = 20. The fences would be:

Lower Fence: 60 - 1.5 * 20 = 30

Upper Fence: 80 + 1.5 * 20 = 110

Any score below 30 or above 110 would be considered a potential outlier.

**Benefits of Using IQR for Outlier Detection**

**Robustness:-** The IQR method is less sensitive to extreme values than other methods based on the mean and standard deviation.

**Simplicity:-** It's relatively easy to calculate and understand.

**Widely Applicable:-** It can be used for a variety of data types and distributions.

**Considerations**

The 1.5 * IQR multiplier is a common convention, but it can be adjusted depending on the specific dataset and the desired level of sensitivity to outliers.

Outliers identified by the IQR method should be investigated further to determine if they are genuine errors or simply unusual but valid data points.

**Conclusion**

The interquartile range (IQR) is a useful measure of statistical dispersion that can be used to detect potential outliers in a dataset. By establishing fences based on the IQR, we can identify data points that are significantly far from the central 50% of the data, warranting further investigation. The IQR method is robust, simple, and widely applicable, making it a valuable tool for data analysis.

###8. Discuss the conditions under which the binomial distribution is used.

Ans:- The binomial distribution is a discrete probability distribution that describes the probability of obtaining a certain number of successes in a sequence of independent trials, where each trial has only two possible outcomes (success or failure).

Here are the conditions under which the binomial distribution is used:-

**1.Fixed Number of Trials (n):-** There must be a predetermined, fixed number of trials. For example, flipping a coin 10 times or surveying 100 people.

**2.Independent Trials:-** Each trial must be independent of the others. The outcome of one trial should not affect the outcome of any other trial. For example, if you flip a coin multiple times, the outcome of one flip doesn't change the probability of heads or tails on the next flip.

**3.Two Possible Outcomes:-** Each trial must have only two possible outcomes: success or failure. These outcomes are mutually exclusive, meaning they cannot occur at the same time. For example, a coin flip can result in heads (success) or tails (failure).

**4.Constant Probability of Success (p):-** The probability of success must remain the same for each trial. For example, if the probability of getting heads on a coin flip is 0.5, it should remain 0.5 for all flips.

When these conditions are met, we can use the binomial distribution to calculate the probability of getting a specific number of successes (k) in n trials.

**Formula:-**

The probability mass function of the binomial distribution is given by:

    P(X = k) = (n choose k) * p^k * (1 - p)^(n - k)

Where:-

P(X = k) is the probability of getting exactly k successes in n trials.

(n choose k) is the binomial coefficient, which represents the number of ways to choose k successes from n trials.

p is the probability of success on a single trial.
(1 - p) is the probability of failure on a single trial.

**Example:-**

Consider flipping a fair coin 5 times (n = 5). The probability of getting heads (success) on a single flip is 0.5 (p = 0.5). We want to calculate the probability of getting exactly 3 heads (k = 3).

Using the binomial distribution formula:-


P(X = 3) = (5 choose 3) * (0.5)^3 * (0.5)^(5 - 3)
         = 10 * (0.125) * (0.25)
         = 0.3125

Therefore, the probability of getting exactly 3 heads in 5 flips is 0.3125 or 31.25%.

###9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

ANS:- **Properties of the Normal Distribution**

The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is widely used in statistics and data science. It is characterized by its bell-shaped curve, which is symmetric around the mean.

Here are the key properties of the normal distribution:-

**1.Bell-Shaped and Symmetric:-** The distribution is bell-shaped and symmetric around its mean (µ). This means that the left and right halves of the curve are mirror images of each other.

**2.Mean, Median, and Mode are Equal:-** The mean, median, and mode of a normal distribution are all equal and located at the center of the curve.

**3.Empirical Rule:-** The empirical rule, also known as the 68-95-99.7 rule, states that approximately:

(a)68% of the data falls within one standard deviation (σ) of the mean.

(b)95% of the data falls within two standard deviations of the mean.

(c)99.7% of the data falls within three standard deviations of the mean.

**4.Total Area Under the Curve is 1:-** The total area under the normal distribution curve is equal to 1, representing the total probability of all possible outcomes.

**5.Defined by Mean and Standard Deviation:-** The normal distribution is completely defined by its mean (µ) and standard deviation (σ). These two parameters determine the location and shape of the curve.

**Empirical Rule (68-95-99.7 Rule)**

The empirical rule is a useful heuristic for understanding the spread of data in a normal distribution. It provides a quick way to estimate the percentage of data that falls within certain ranges around the mean.

Here's a breakdown of the rule:-

**1. 68% of data within 1 standard deviation of the mean:-** This means that approximately 68% of the data points in a normal distribution will fall within the range of (µ - σ) to (µ + σ).

**2. 95% of data within 2 standard deviations of the mean:-** This means that approximately 95% of the data points will fall within the range of (µ - 2σ) to (µ + 2σ).

**3. 99.7% of data within 3 standard deviations of the mean:-** This means that approximately 99.7% of the data points will fall within the range of (µ - 3σ) to (µ + 3σ).

**Example:-**

Let's say the height of adult women is normally distributed with a mean of 165 cm and a standard deviation of 5 cm. Using the empirical rule, we can estimate that:

Approximately 68% of adult women have heights between 160 cm (165 - 5) and 170 cm (165 + 5).

Approximately 95% of adult women have heights between 155 cm (165 - 25) and 175 cm (165 + 25).

Approximately 99.7% of adult women have heights between 150 cm (165 - 35) and 180 cm (165 + 35).

**Importance of the Normal Distribution and Empirical Rule**

The normal distribution is a fundamental concept in statistics and is used in many statistical tests and models. The empirical rule provides a quick and easy way to understand the spread of data in a normal distribution, making it a valuable tool for data analysis and interpretation.

###10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.

ANS:- **Real-life Example: Customer Arrivals at a Store**

Consider a retail store where customers arrive randomly throughout the day. The number of customers arriving within a specific time interval can be modeled as a Poisson process.

**Assumptions:-**

Customers arrive independently of each other.

The average rate of customer arrivals is constant over time.

The probability of two or more customers arriving at the exact same time is negligible.

**Scenario:-**

Let's say the store experiences an average of 10 customer arrivals per hour (λ = 10). We want to calculate the probability of exactly 5 customers arriving in the next hour.

**Calculation:-**

The probability mass function of the Poisson distribution is given by:

    P(X = k) = (e^-λ * λ^k) / k!

Where:

P(X = k) is the probability of observing exactly k events in a given time interval.

e is the base of the natural logarithm (approximately 2.71828).

λ is the average rate of events (in this case, customer arrivals per hour).

k is the number of events we want to calculate the probability for.

In our scenario, we have λ = 10 and k = 5. Plugging these values into the formula:


    P(X = 5) = (e^-10 * 10^5) / 5!
             ≈ (0.0000454 * 100000) / 120
             ≈ 0.0378

Therefore, the probability of exactly 5 customers arriving in the next hour is approximately 0.0378 or 3.78%.

**Interpretation:-**

This means that there is a 3.78% chance that exactly 5 customers will arrive at the store in the next hour, given the average arrival rate of 10 customers per hour.

**Other Applications of Poisson Processes:-**

Number of cars passing a certain point on a highway in a given time interval.

Number of phone calls received at a call center per hour.

Number of defects in a manufactured product.

Number of radioactive decay events in a given time period.

###11. Explain what a random variable is and differentiate between discrete and continuous random variables.

Ans:-  In probability and statistics, a random variable is a variable whose value is a numerical outcome of a random phenomenon. It is a function that maps the outcomes of a random process to numerical values.

**Think of it this way:-**

Imagine we're flipping a coin. The outcome of a single flip (heads or tails) is random. We can define a random variable, say X, to represent the number of heads obtained in a series of coin flips.

If we flip the coin once, X can take the values 0 (tails) or 1 (heads).

If we flip the coin twice, X can take the values 0, 1, or 2.

So, the random variable X assigns numerical values to the possible outcomes of the coin flips.

**Types of Random Variables:-** There are two main types of random variables:-

**1. Discrete Random Variables:-** A discrete random variable can only take on a finite or countably infinite number of distinct values. These values are usually integers, but the total number of values must be finite or countable.

**Examples:-**

The number of heads in a series of coin flips.

The number of cars passing a certain point in an hour.
The number of students in a class.

**2.Continuous Random Variables:-** A continuous random variable can take on any value within a given range or interval. There are infinitely many possible values that the variable can take.

**Examples:-**

The height of a person.

The temperature of a room.

The weight of an object.

**Key Differences**

    Feature	  Discrete Random Variable	      Continuous Random Variable

    Values	   Finite or countably               Any value within a range
                 infinite distinct values

    Type of data      Count data	                Measurement data

    Probability  Probability mass function (PMF)	Probability density function (PDF)

    Graph	    Represented by a bar graph	     Represented by a smooth curve

**In summary:-**

**Discrete random variables** are used for counting and representing data that can only take on specific, separate values.

**Continuous random variables** are used for measuring and representing data that can take on any value within a range.

###12.  Provide an example dataset, calculate both covariance and correlation, and interpret the results.

Ans:- **Example Dataset**

Let's consider a dataset with two variables:-

**X:-** Hours of study

**Y:-** Exam scores

    Hours of Study (X)	Exam Score (Y)

       2	                  60
       4	                  70
       6	                  80
       8	                  90
       10	                 100

**Calculating Covariance**

Covariance measures the directional relationship between two variables. A positive covariance indicates that the variables tend to move in the same direction, while a negative covariance indicates that they tend to move in opposite directions.

Here's the formula for calculating covariance:-


    Cov(X, Y) = Σ[(Xi - X̄)(Yi - Ȳ)] / (n - 1)

Where:-

Xi and Yi are individual data points

X̄ and Ȳ are the means of X and Y, respectively
n is the number of data points

Applying this formula to our dataset:-


    X̄ = (2 + 4 + 6 + 8 + 10) / 5 = 6

    Ȳ = (60 + 70 + 80 + 90 + 100) / 5 = 80

    Cov(X, Y) = [(2-6)(60-80) + (4-6)(70-80) + (6-6)(80-80) + (8-6)(90-80) + (10-6)(100-80)] / (5-1)

              = 200 / 4
              = 50

**Calculating Correlation**

Correlation measures the strength and direction of the linear relationship between two variables. It's a standardized version of covariance, ranging from -1 to +1.

Here's the formula for calculating correlation:-


    Corr(X, Y) = Cov(X, Y) / (σX * σY)

Where:-

Cov(X, Y) is the covariance between X and Y

σX and σY are the standard deviations of X and Y, respectively

First, we need to calculate the standard deviations:-


    σX = √[Σ(Xi - X̄)² / (n - 1)] = √[(2-6)² + (4-6)² + (6-6)² + (8-6)² + (10-6)²] / (5-1) = 2.83

    σY = √[Σ(Yi - Ȳ)² / (n - 1)] = √[(60-80)² + (70-80)² + (80-80)² + (90-80)² + (100-80)²] / (5-1) = 14.14

    Corr(X, Y) = 50 / (2.83 * 14.14)
               ≈ 1

**Interpretation**

**Covariance:-** The positive covariance of 50 indicates that as the hours of study (X) increase, the exam scores (Y) also tend to increase. There's a positive relationship between the two variables.

**Correlation:-** The correlation of 1 indicates a perfect positive linear relationship between hours of study and exam scores. This means that as hours of study increase, exam scores increase proportionally.

In this example, we see a strong positive relationship between the two variables. More hours of study are associated with higher exam scores.
